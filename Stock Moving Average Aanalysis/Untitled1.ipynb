```python
import pandas as pd
import yfinance as yf
# Define the list of stock symbols
stock_symbols = ['AAPL', 'GOOGL', 'TSLA']  # Add more symbols as needed

# Fetch the historical data for the stocks
data = yf.download(stock_symbols, start='2023-01-01', end='2023-12-31')
# Calculate the 30-day average volume for each stock
data['30-day Average Volume'] = data['Volume'].rolling(window=30).mean()
# Filter the DataFrame to select stocks with volume > 30-day average volume
filtered_data = data[data['Volume'] > data['30-day Average Volume']]
filtered_data
```
