```python
def on_ticks(ticks):
    print(ticks)
import time
import pandas as pd
from datetime import timedelta, datetime

    
breeze.ws_connect()
breeze.on_ticks = on_ticks
    
# breeze.subscribe_feeds(stock_token = "i_click_2_gain")
def get_sma():
    STOCK = "TCS"
    current_time = datetime.now()
    from_date = current_time - timedelta(minutes=4)
    
    # note that we have converted dates to ISO time-format before making API call
    data = breeze.get_historical_data(interval='1minute',
                                        from_date = from_date.strftime('%Y-%m-%dT%H:%M:%S.000Z'),
                                        to_date = current_time.strftime('%Y-%m-%dT%H:%M:%S.000Z'),
                                        stock_code=STOCK,
                                        exchange_code="NSE",
                                        product_type="cash")


    # Calculate simple moving average of 'close' price
    print(data)
    data = pd.DataFrame(data['Success'])
    data.close = data.close.astype(float)
    sma = round(data.close.mean(),2)
    return sma
indicator = get_sma()
print(indicator)
```
